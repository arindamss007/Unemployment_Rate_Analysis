***UNEMPLOYMENT RATE ANALYSIS USING PYTHON***

**NOTE** : These analysis are based on past data 2019-20 during covid-19 situation. In pandamic Situation, the employment rates are a good measure for forthcoming pandamic issues like covid-19 occurs.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Unemployment_in_India.csv")
df.dropna(inplace=True)
df
```

```python
X = df['Area']
Y = df[' Unemployment Rate']

plt.barh(X,Y)
plt.title("Area Wise Unemployment")
plt.xlabel("Overall Percentage")

plt.show()
```

```python
Y1 = df[' Estimated_Employed']
X1 = df['Region']

plt.barh(X1,Y1)
plt.title("State Wise Employment Rate")
plt.xlabel("Employment Count(1*e7)")
plt.ylabel("States")
plt.show()
```

```python
X2 = df['Region']
Y2 = df[' Unemployment Rate']

plt.barh(X2, Y2)
plt.title("State Wise Unemployment Rate")
plt.xlabel("Overall Percentage")
plt.ylabel("States")
plt.show()
```

```python
x = df['Region']
y = df[' Estimated_Labour_Participation_Rate']

plt.barh(x,y)
plt.title("State Wise Labour Participation Rate")
plt.xlabel("Overall Participation Rate")
plt.ylabel("States")
plt.plot()
```

***UNEMPLOYMENT RATE ANALYSIS - 2020***

**NOTE** : These analysis are based on past data 2020 during covid-19 situation. In pandamic Situation, the employment rates are a good measure for forthcoming pandamic issues like covid-19 occurs.

```python

df2 = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')

df2
```

```python
df2.info()
```

```python
Y1 = df2[' Estimated Employed']
X1 = df2['Region']

plt.barh(X1,Y1)
plt.title("State Wise Employment Rate (2020)")
plt.xlabel("Employment Count(1*e7)")
plt.ylabel("States")
plt.show()
```

```python
X2 = df2['Region']
Y2 = df2[' Estimated Unemployment Rate (%)']

plt.barh(X2, Y2)
plt.title("State Wise Unemployment Rate (2020)")
plt.xlabel("Overall Percentage")
plt.ylabel("States")
plt.show()
```

```python
x = df2['Region']
y = df2[' Estimated Labour Participation Rate (%)']

plt.barh(x,y)
plt.title("State Wise Labour Participation Rate (2020)")
plt.xlabel("Overall Participation Rate")
plt.ylabel("States")
plt.plot()
```

```python
#Building A ML Model for predict Unemployment Rate by providing details of Location, Employed & perticipation rate.

from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

X = df2[['longitude','latitude', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']]
Y = df2[" Estimated Unemployment Rate (%)"]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size=0.80)

emp_model = linear_model.LinearRegression()
emp_model.fit(X_train, Y_train)

predicts = emp_model.predict(X_test)

print("Mean Absolute Error (MAE) : ", mean_absolute_error(predicts, Y_test))
```

**NOTE** : *The Average Rate Of prediction error is very high, if we need to improve the model, we need more data to predict the target with higher accuracy.*
